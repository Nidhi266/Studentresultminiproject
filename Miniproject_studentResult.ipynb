```python
import pandas as pd

# Created a dictonary
A1 = {
    'Name': ['Nidhi', 'Nidhi', 'Patel', 'Zeel'],
    'Marks': [26, 29, 28, 29],
    'Years': ['First_year', 'Final_year', 'Second_year', 'Third_year']
}

# Created the two dimensional data using df
Student = pd.DataFrame.from_dict(A1)

# Grouping by 'Name' on Marks and summing the 'Marks' for each student
split_name = Student.groupby('Name')['Marks'].sum()
print(split_name)


#to calculate the highest marks used idxmax and used location to find the high mark splitname 
highest_marks_index = split_name.idxmax() #Cannot use location here as its string 
print(highest_marks_index,split_name.loc[highest_marks_index])

# Plot pie chart
plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(split_name, labels=None, autopct='%1.1f%%', startangle=140, wedgeprops=dict(edgecolor='w'))
plt.title('Marks obtained by Student', pad=20)
plt.axis('equal')

plt.legend(labels=split_name.index, loc='center left', bbox_to_anchor=(1, 0.5), fontsize='medium')


plt.show()

```
